### VISUALISER DES CATEGORIES

__C'est la gueule typique (TYPIQUE !) d'un dataset avec des categories__ (le sexe ou la classe correspondent a des categories 0, 1, 2, etc.)

C'est pourquoi il va falloir utiliser des fonctions de [Categorical plot](https://seaborn.pydata.org/api.html#categorical-plots)

```python
sns.catplot(x = 'pclass', y = 'age', data=titanic, hue='sex')
```

```python
sns.boxplot(x = 'pclass', y = 'age', data=titanic, hue='sex')
```

### VISUALISER DES REPARTITIONS

```python
sns.displot(titanic['fare'])
```

N.B : La fonction [jointplot](https://seaborn.pydata.org/generated/seaborn.jointplot.html#seaborn.jointplot) permet de voir la distribution selon deux categories

```python
sns.jointplot(x = 'age', y = 'fare', data=titanic, kind = 'kde') # kernel density estimation / ou hex pour hexagonal
```

```python
sns.heatmap(titanic.corr())
```
